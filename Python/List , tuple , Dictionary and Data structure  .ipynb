## Part A – Basic Tasks (Lists)

### Task A1: Print each element using a loop

```python
numbers = [10, 20, 30, 40, 50]
for num in numbers:
    print(num)
```

### Task A2: Access first, last, and middle element

```python
data = [5, 15, 25, 35, 45]
print('First:', data[0])
print('Last:', data[-1])
print('Middle:', data[2])
```

### Task A3: Reverse a list using slicing

```python
items = [1, 2, 3, 4, 5]
reversed_items = items[::-1]
print(reversed_items)
```

### Task A4: Add and remove elements

```python
fruits = ['apple', 'banana', 'cherry']
fruits.append('mango')
fruits.insert(1, 'grape')
fruits.remove('banana')
fruits.pop()
print(fruits)
```

### Task A5: Find maximum and minimum value

```python
values = [34, 7, 89, 12, 56]
print('Max:', max(values))
print('Min:', min(values))
```

## Part B – Intermediate Tasks (Lists & Tuples)

### Task B1: Tuple modification (immutability)

```python
t = (1, 2, 3)
print(t)
# Tuples are immutable. The line below will raise a TypeError:
# t[0] = 99
```

### Task B2: Count even and odd numbers

```python
nums = [3, 8, 15, 22, 7, 4]
even = sum(1 for x in nums if x % 2 == 0)
odd = len(nums) - even
print('Even:', even, '| Odd:', odd)
```

### Task B3: Nested list (matrix) and element access

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print('Element at row 1, col 2:', matrix[1][2])
```

### Task B4: Sort list ascending and descending

```python
scores = [42, 15, 78, 31, 90]
print('Ascending:', sorted(scores))
print('Descending:', sorted(scores, reverse=True))
```

### Task B5: List comprehension (squares)

```python
squares = [x**2 for x in range(1, 6)]
print(squares)
```

## Part D – Mini Project: Student Record Management

```python
records = {}

records['Ali'] = 85
records['Umer'] = 91
records['Usman'] = 73
records['areeb'] = 88

print('--- All Records ---')
for name, marks in records.items():
    print(f'{name}: {marks}')

search = 'Sara'
print('\nSearch:', search, '->', records.get(search, 'Not found'))

print('\n--- Sorted by Marks (High to Low) ---')
for name, marks in sorted(records.items(), key=lambda x: x[1], reverse=True):
    print(f'{name}: {marks}')
```

## Part E – Conceptual Questions

**Q1:** List is mutable (can be changed); Tuple is immutable (cannot be changed).

**Q2:** Dictionaries store data as key-value pairs, making search and retrieval fast and easy.

**Q3:** Mutability means you can add, remove, or modify elements in a list after creation.

**Q4:** Indexing accesses a single element (e.g., `list[0]`). Slicing extracts a portion (e.g., `list[1:4]`).

**Q5:** `sort()` modifies the original list in place. `sorted()` returns a new sorted list without changing the original.
