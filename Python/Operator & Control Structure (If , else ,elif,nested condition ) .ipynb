# **Part B: NESTED CONDITION**

**Task B1:**

```python
age = int(input("Enter your age: "))
nationality = input("Enter your nationality: ")

if age >= 18:
    if nationality.lower() == "pakistani":
        print("Eligible to Vote")
    else:
        print("Not Eligible")
else:
    print("Not Eligible")
```

**Task B2:**

```python
marks = int(input("Enter your marks: "))
interview = input("Interview result (pass/fail): ")

if marks >= 70:
    if interview.lower() == "pass":
        print("Admission Granted")
    else:
        print("Admission Denied")
else:
    print("Admission Denied")
```

**Task B3:**

```python
distance = float(input("Enter distance in km: "))

if distance < 1:
    print("Walk")
elif distance <= 5:
    print("Use Bike")
else:
    print("Use Bus or Car")
```

# **Part C: Conceptual and Real‑World Thinking Tasks**

***Conceptual Task 1:***


In my daily life, I make many decisions using conditions. For example, **if** I **wake up late**, I skip **breakfast**, otherwise **I eat it**. **If **I have **homework**, I stay at home, **else** I go outside. Also, **if** my **mobile battery is low**, I charge it. These are like **if–else **conditions in programming.

***Conceptual Task 2:***

A computer can make wrong decisions when it does not understand real-life situations. **For example**, if a student is late because of an **emergency**, the system will still mark them **absent**. But a teacher can understand the reason and allow them. So computers cannot understand every situation **like humans**.

***Conceptual Task 3:***

In **university**, there is a rule for exams. **If** a student has more than **80% **attendance, they can sit in **exams**, otherwise they cannot. This rule can be written using an **if–else** condition in programming.

***Conceptual Task 4:***

A real-life example of **nested conditions** is a **police test**. First, written test marks are **checked**. **If** marks are good, then physical test is checked. **If** the candidate passes both, then they are selected. One condition is not enough, **so multiple conditions are needed**

***Conceptual Task 5:***

Sometimes human thinking is important even if a system gives advice. For example, **Google Maps** shows the shortest path, but it may have** traffic **or **bad roads**. A human can decide **a better** way. So human judgment is **still important**
