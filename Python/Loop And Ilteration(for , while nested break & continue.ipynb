# **Part E: Conceptual Tasks**

***Conceptual Task 1:***

**1. Importance of loops in AI:**
Loops allow AI/ML programs to repeat tasks automatically, like training a model over multiple epochs or iterating through datasets.

**Example:** A loop trains a neural network 50 times over the training data to improve accuracy.

***Conceptual Task 2:***

**2. Difference between for and while loops:**

**For loop:** Runs a fixed number of times.

**Example:** Iterating through 1000 images in a dataset for one epoch.

**While loop:** Runs until a condition is met.

**Example:** Keep training the model until the validation loss stops decreasing.

***Conceptual Task 3:***

**3. Use of nested loops in real applications:**
Nested loops help process complex data structures.

**Example:**

Loop over images → loop over each pixel → apply a filter in convolutional neural networks (CNNs).

Generating all possible feature combinations in feature selection.

***Conceptual Task 4:***

**4. What happens in infinite loops?**

Infinite loops happen when the stop condition is never reached.

**Example:** Training a model without a proper stopping criterion could run endlessly, consuming CPU/GPU and memory.

# **Part F: Human Intelligence Tasks**

**F1**

**1. Identify repetitive tasks in daily life (like ML iterations):**

Checking emails repeatedly → like repeatedly updating weights in gradient descent
Brushing teeth daily → like running multiple epochs on training data

**F2**

**Repetition is harmful** if done excessively without improvement.

**Example**: Overtraining a model on the same data leads to overfitting → poor performance on new data.

**F3**

**3. Example of Early Stopping:**

Early stopping refers to halting a process at an optimal point to prevent overexertion or inefficiency.

**Examples:**

Ending a training session upon completing **planned exercises.**

**In machine learning,** stopping model training when validation accuracy stops improving to prevent overfitting.

**F4**

**4. Example of Skipping Steps:**

Skipping steps means omitting necessary actions to save time, which can compromise results.

**Examples:**

Preparing a procedure without following all required instructions.
**In machine learning,** skipping data preprocessing can lead to poor model performance
