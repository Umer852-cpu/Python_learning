```python
from datetime import date
year, month, day = map(int, input("Enter DOB (YYYY MM DD): ").split())
today = date.today()
years = today.year - year
months = today.month - month
days = today.day - day
if days < 0:
    months -= 1
    days += 30
if months < 0:
    years -= 1
    months += 12
print(f"{years} years {months} months {days} days")
```

```python
i = 1

while i <= 5:
    sub = int(input("Enter marks for subject " + str(i) + ": "))
    gpa = max(0, min(4, 1 + (sub - 50) / 10))
    print("GPA for subject", i, ":", round(gpa, 2))
    i += 1
```

```python
i = 1
total_gpa = 0
while i <= 5:
    sub = int(input("Enter marks for subject " + str(i) + ": "))
    gpa = max(0, min(4, 1 + (sub - 50) / 10))
    print("GPA for subject", i, ":", round(gpa, 2))
    total_gpa = total_gpa + gpa

    i += 1

sgpa = total_gpa / 5

print("Your SGPA is:", round(sgpa, 2))
```
