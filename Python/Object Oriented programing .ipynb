## Part B

**B1 — BankAccount with private balance**

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance   # private

    def deposit(self, amt):
        self.__balance += amt
        print(f"Deposited {amt}. Balance: {self.__balance}")

    def withdraw(self, amt):
        if amt <= self.__balance:
            self.__balance -= amt
            print(f"Withdrawn {amt}. Balance: {self.__balance}")
        else:
            print("Low balance!")

    def balance(self):
        return self.__balance

acc = BankAccount("Ali", 10000)
acc.deposit(2000)
acc.withdraw(5000)
print("Final:", acc.balance())
```

**B2 — Vehicle → Car inheritance**

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    def show(self):
        print(f"{self.brand} | {self.speed} km/h")

class Car(Vehicle):
    def __init__(self, brand, speed, doors):
        super().__init__(brand, speed)
        self.doors = doors
    def show(self):
        super().show()
        print(f"Doors: {self.doors}")

Car("Honda", 160, 4).show()
```

**B3 — Method overriding**

```python
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):            # overrides parent
        print("Woof!")

Animal().speak()
Dog().speak()
```

**B5 — LibraryBook**

```python
class LibraryBook:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.issued = False

    def issue(self, student):
        if not self.issued:
            self.issued = True
            print(f"'{self.title}' issued to {student}")
        else:
            print("Already issued")

    def return_book(self):
        if self.issued:
            self.issued = False
            print(f"'{self.title}' returned")
        else:
            print("Not issued")

b = LibraryBook("AI Basics", "Russell")
b.issue("Umer")
b.issue("Areeb")
b.return_book()
b.issue("Umet")
```

**C1 — AI Agent**

```python
class AIAgent:
    def __init__(self, name):
        self.name = name

    def observe(self, env):
        print(f"[{self.name}] Observing: {env}")

    def act(self, action):
        print(f"[{self.name}] Acting: {action}")

a = AIAgent("Agent-X")
a.observe("High temperature detected")
a.act("Activate cooling system")
```

**C2 — Chatbot**

```python
class Chatbot:
    def __init__(self, name):
        self.name = name

    def respond(self, msg):
        m = msg.lower()
        if "hi" in m or "hello" in m:
            print(f"{self.name}: Hi! How can I help?")
        elif "bye" in m:
            print(f"{self.name}: Goodbye!")
        else:
            print(f"{self.name}: I only understand greetings.")

bot = Chatbot("PyBot")
bot.respond("Hello!")
bot.respond("What is Python?")
bot.respond("Bye")
```

**C5 — PredictionModel**

```python
class PredictionModel:
    def __init__(self, name):
        self.name = name
        self.trained = False

    def train(self, data):
        self.trained = True
        print(f"{self.name} trained on {len(data)} samples")

    def predict(self, val):
        if not self.trained:
            print("Train the model first!")
        else:
            label = "Pass" if val >= 50 else "Fail"
            print(f"Prediction for {val}: {label}")

m = PredictionModel("Model-A")
m.predict(70)            # not trained
m.train([60,70,80,45,90])
m.predict(70)
m.predict(30)
```

**D1 — Three real-world OOP systems**

1. **Hospital System** — `Patient`, `Doctor`, `Appointment` classes. Doctors treat patients, appointments are linked to both. Each has clear attributes and behaviors.

2. **University System** — `Student`, `Teacher`, `Course`, `Department`. Enrollment, teaching, grading — all modeled as object interactions.

3. **E-Commerce Store** — `Product`, `Cart`, `Order`, `Customer`. A customer picks products, adds to cart, and checks out. Clean separation of responsibilities.


**D2 — Inheritance in university management**

Different roles like `Student`, `Teacher`, and `Admin` all share common data: name, ID, contact. Inheritance lets us define these once in a `Person` base class and reuse them across all roles.

Adding a new field to `Person` automatically updates all subclasses. This keeps code clean, avoids repetition, and reflects the actual hierarchy in the real institution.


**D3 — Encapsulation for privacy**

A bank account is the clearest example. Your `__balance` is private — you cannot do `account.balance = 1000000` from outside. The class only exposes safe methods like `deposit()` and `withdraw()` which validate input before touching the balance.

This is exactly how real banks work — controlled access, not direct access.


**D4 — Human assistant vs AI Agent**

| | Human Assistant | AI Agent |
|---|---|---|
| Attributes | name, skills, mood | name, version, state |
| observe() | senses, perception | input data / sensors |
| act() | physical/verbal | code execution |
| learn() | experience | retraining on data |

Both map naturally to OOP objects. The AI agent is more predictable but less flexible than a human.


**D5 — My own class: SmartFan**

```python
class SmartFan:
    def __init__(self, brand):
        self.brand = brand
        self.speed = 1
        self.is_on = False

    def turn_on(self):
        self.is_on = True
        print(f"{self.brand} ON")

    def turn_off(self):
        self.is_on = False
        print(f"{self.brand} OFF")

    def set_speed(self, n):
        if 1 <= n <= 5:
            self.speed = n
            print(f"Speed: {n}")
        else:
            print("Speed 1-5 only")

    def status(self):
        state = "ON" if self.is_on else "OFF"
        print(f"{self.brand} | {state} | Speed {self.speed}")

f = SmartFan("Orient")
f.turn_on()
f.set_speed(4)
f.status()
```

## Part E — Mini Project: Student Management System

```python
class Student:
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks

    def display(self):
        g = "A" if self.marks>=90 else "B" if self.marks>=80 else "C" if self.marks>=70 else "D"
        print(f"[{self.roll}] {self.name} — {self.marks} ({g})")

    def update(self, name=None, marks=None):
        if name: self.name = name
        if marks is not None: self.marks = marks

# list of students
students = [
    Student("Ali",   101, 88),
    Student("Sara",  102, 95),
    Student("Bilal", 103, 72),
    Student("Hina",  104, 60),
]

print("-- All Students --")
for s in students: s.display()

students[0].update(marks=92)
print()
print("-- After Update --")
for s in students: s.display()
```

## Part F — Conceptual Questions

**1. Class vs Object**
A class is the blueprint. An object is an actual instance. `Car` is the class; `Car("Toyota", 180)` is the object.

**2. Why OOP?**
Organizes code, avoids repetition, models real world naturally, makes large programs easier to manage and extend.

**3. Constructor**
`__init__()` in Python. Runs automatically when an object is created. Sets up initial attribute values.

**4. Encapsulation**
Hiding data inside a class using private variables (`__var`). Only allowed access through controlled methods. Protects data from direct modification.

**5. Inheritance vs Overriding**
- **Inheritance** — child class gets all parent methods and attributes
- **Overriding** — child class redefines a parent method to change its behavior
