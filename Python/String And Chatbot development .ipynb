## Part C: Text Analyzer Project

```python
text = input("Enter text: ")

# count words
word_count = len(text.split())

# count vowels
vowel_count = 0
for ch in text:
    if ch in "aeiouAEIOU":
        vowel_count += 1

# count digits
digit_count = 0
for ch in text:
    if ch.isdigit():
        digit_count += 1

# count special characters (not letter, not digit, not space)
special_count = 0
for ch in text:
    if not ch.isalnum() and ch != " ":
        special_count += 1

print("Words:", word_count)
print("Vowels:", vowel_count)
print("Digits:", digit_count)
print("Special characters:", special_count)
```

## Part D: Chatbot Development

```python
print("Chatbot started. Type 'bye' to exit.")

while True:
    user = input("You: ").lower()

    if "hello" in user or "hi" in user:
        print("Bot: Hello! How can I help you?")

    elif "what is ai" in user or "artificial intelligence" in user:
        print("Bot: AI means making computers think and learn like humans.")

    elif "python" in user:
        print("Bot: Python is a programming language used a lot in AI and data science.")

    elif "course" in user:
        print("Bot: This course is about programming for artificial intelligence.")

    elif "bye" in user:
        print("Bot: Goodbye! Have a nice day.")
        break

    elif "joke" in user:
        print("Bot: Why do programmers hate nature? It has too many bugs.")

    elif "help" in user:
        print("Bot: I can answer questions about AI, Python, or this course.")

    else:
        print("Bot: Sorry, I did not understand that.")
```

## **Part E: Conceptual Questions**



**Q1: Why are strings important in AI and NLP?**

Strings are important because all human language is text. When we build AI systems that understand language, everything starts as a string. Tasks like chatbots, translation, and sentiment analysis all work with strings as the main input.

**Q2: Explain difference between indexing and slicing.**

Indexing is used to get a single character from a string by its position. For example s[0] gives the first character. Slicing is used to get a part of the string using a start and end position like s[1:4] which gives characters from index 1 to 3.

**Q3: What is string immutability?**

String immutability means that once a string is created in Python you cannot change it directly. If you try to do something like s[0] = 'A' it gives an error. Any operation that modifies a string actually creates a new string.

**Q4: Why do we use lower() in chatbot systems?**

We use lower() so the chatbot can match user input without worrying about uppercase or lowercase letters. If a user types "Hello" or "HELLO" or "hello" they all become "hello" after lower() and the same condition works for all of them.

**Q5: Difference between split() and join().**

split() breaks a string into a list of words. For example "hello world".split() gives ['hello', 'world']. join() does the opposite, it takes a list and combines it into a string. For example " ".join(['hello', 'world']) gives "hello world".

## **Part F: Human Intelligence Tasks**



**Task F1: How a chatbot mimics conversation**

Humans talk by listening, understanding the meaning, and then replying. A chatbot does something similar but in a simpler way. It takes the user's input, looks for keywords in it, and sends back a pre-written response that matches those keywords. It does not truly understand but it gives replies that feel like a real conversation.

**Task F2: Where a chatbot may fail to understand emotion**

If someone types "I am fine I guess" the chatbot might think they are happy because of the word fine. But actually the person might be sad or confused. The chatbot cannot tell the difference because it only looks at words and not the feeling behind them.

**Task F3: Different responses for the same question**

For the question "what is AI" you could answer differently depending on who is asking. For a child you might say it means making computers smart. For a student you might give a proper definition. For an expert you might talk about neural networks and machine learning algorithms. The same question can have many valid answers depending on the audience.

**Task F4: Real-world application of chatbot**

Many companies use chatbots in customer service. For example an online shopping website uses a chatbot to answer questions about order tracking, returns, and delivery times. This saves time for both the customer and the company because the customer gets an instant reply without waiting for a human agent.
