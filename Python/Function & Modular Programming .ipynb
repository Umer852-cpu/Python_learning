## Part A — Basic Function Tasks

### **Task 1**


````python
def square(n):
    return n * n

print(square(5))
print(square(9))
````

**Why are functions important?**

Functions let us wrap a piece of logic once and reuse it wherever we need it. Instead of writing `n * n` every single time I need a square, I just call `square(n)`. It keeps the code clean and easy to debug too — if there's a bug, I fix it in one place instead of hunting through the whole file.

### **Task 2**

````python
def add_two(a, b):
    return a + b

result = add_two(10, 25)
print(result)
````

**Difference between `return` and `print`:**

- `print` just shows the value on screen — the function doesn't actually send anything back. You can't store it.
- `return` gives the value back to whoever called the function. So `result = add_two(10, 25)` actually stores 35 in `result`.

In real code you almost always want `return`, because you usually need to use that value somewhere else.

## **Part B — Intermediate Tasks (Parameters & Return Values)**

### **Task B1**

````python
def power(base, exp=2):   # default exponent is 2
    return base ** exp

print(power(3))      # uses default, gives 9
print(power(3, 3))   # overriding default, gives 27
print(power(2, 10))  # 1024
````

### **Task B2**

````python
def total_sum(*nums):
    return sum(nums)

print(total_sum(1, 2, 3))          # 6
print(total_sum(5, 10, 15, 20))    # 50
print(total_sum(100, 200))         # 300
````

### **Task B3**

````python
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(find_max(10, 45, 30))   # 45
print(find_max(7, 7, 3))      # 7
print(find_max(1, 1, 1))      # 1
````

### **Task B4**

````python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in [2, 3, 4, 17, 20, 29]:
    print(f"{num} -> Prime: {is_prime(num)}")
````

### **Task B5**

````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print(factorial(0))   # 1
print(factorial(5))   # 120
print(factorial(7))   # 5040
````

### **Task C1**

````python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: can't divide by zero"
    return a / b

# quick test
print(add(10, 5))
print(subtract(10, 5))
print(multiply(10, 5))
print(divide(10, 5))
print(divide(10, 0))
````

## Part E — Conceptual Questions

**Q1: Why are functions important in AI systems?**

AI systems involve a lot of repeated operations — preprocessing data, making predictions, calculating loss, updating weights, etc. Functions let you define each of these steps once and call them as many times as needed. Without functions, training a model would mean copy-pasting the same logic thousands of times, which is both unmanageable and bug-prone. Functions also make it easy to swap one part of the system (e.g., change the loss function) without touching everything else.

**Q2: Difference between parameters and arguments**

- **Parameters** are the variable names in the function definition. They're basically placeholders.
- **Arguments** are the actual values you pass when calling the function.

```python
def greet(name):      # 'name' is a parameter
    print("Hi", name)

greet("Ahmed")        # "Ahmed" is the argument
```

**Q3: What is variable scope? Local vs Global**

Scope refers to where a variable is accessible in your code.

- **Local variable**: created inside a function, only exists while that function runs. Once the function finishes, it's gone.
- **Global variable**: created outside all functions, accessible from anywhere in the file.

Example:
```python
x = 100   # global

def show():
    y = 50   # local — only exists here
    print(x)  # can access global x
    print(y)

show()
# print(y)  # would crash — y doesn't exist here
```

**Q4: Why is modular programming important?**

When a program grows large, keeping all code in one block makes it a nightmare to maintain. Modular programming means breaking the program into small, self-contained functions or modules. Each one does one thing well. This makes it easier to:
- Test individual pieces separately
- Reuse code across projects
- Work in teams (different people work on different modules)
- Find and fix bugs quickly

**Q5: Difference between `return` and `print`**

| | `return` | `print` |
|---|---|---|
| Sends value back? | Yes | No |
| Can store result? | Yes | No |
| Visible on screen? | Not automatically | Yes |

`print` is for displaying. `return` is for passing a value back so it can actually be used in the program.

## Part F — Human Intelligence Tasks

**Task F1: Real-life tasks that can be broken into smaller functions**

Making tea:
- `boil_water()`
- `add_tea_leaves()`
- `add_milk()`
- `add_sugar()`
- `pour_in_cup()`

Each step is separate and could be skipped or changed (e.g., no sugar) without affecting the others. This is exactly how functions work.

**Task F2: Real-world system where modular design improves efficiency**

A hospital management system is a great example. It can be broken into modules:
- Patient Registration
- Doctor Scheduling
- Billing
- Pharmacy
- Lab Results

Each module works independently. The billing team can update payment logic without touching the lab module. If there's a bug in scheduling, it doesn't crash the pharmacy system. This is exactly what modular design achieves.

**Task F3: Example where function reuse saves time and effort**

Imagine I write a `validate_email(email)` function that checks if an email is in the right format. I can now use that exact function in:
- The registration form
- The login form
- The "forgot password" page
- The admin panel

Without the function, I'd write the same validation logic 4 times. If I ever need to change the validation rule, I change it once and it applies everywhere automatically.

**Task F4: Situation where wrong use of global variables causes problems**

Say you have a global variable `count = 0` that tracks how many items are in a shopping cart. If multiple functions accidentally modify it (one adds to it, another resets it mid-calculation), the cart ends up with wrong numbers. This is a classic bug that's hard to trace because the variable is being touched from all over the place.

The fix is to pass `count` as a parameter or keep it local to the relevant function, so you always know who changed it.

## Mini Project — Menu-Driven Program using Functions

````python
# --- individual operation functions ---

def get_square(n):
    return n * n

def get_cube(n):
    return n ** 3

def check_even_odd(n):
    return "Even" if n % 2 == 0 else "Odd"

def check_pos_neg(n):
    if n > 0:
        return "Positive"
    elif n < 0:
        return "Negative"
    else:
        return "Zero"


# --- main menu ---

def show_menu():
    print("\n==============================")
    print("      Mini Calculator Menu    ")
    print("==============================")
    print("1. Square of a number")
    print("2. Cube of a number")
    print("3. Even or Odd")
    print("4. Positive or Negative")
    print("5. Exit")
    print("==============================")

def run_program():
    while True:
        show_menu()
        choice = input("Enter your choice (1-5): ").strip()

        if choice == '5':
            print("Exiting... Goodbye!")
            break

        if choice not in ['1', '2', '3', '4']:
            print("Invalid choice, try again.")
            continue

        num = float(input("Enter a number: "))

        if choice == '1':
            print(f"Square of {num} = {get_square(num)}")
        elif choice == '2':
            print(f"Cube of {num} = {get_cube(num)}")
        elif choice == '3':
            print(f"{int(num)} is {check_even_odd(int(num))}")
        elif choice == '4':
            print(f"{num} is {check_pos_neg(num)}")

run_program()
````
