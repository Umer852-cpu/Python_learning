# **Part B: Data Type Exploration**

**Task B1:**

```python

num = 10          # integer
price = 25.5      # float
name = "Umer"     # string
is_student = True # boolean

print("Value:", num, "Type:", type(num))
print("Value:", price, "Type:", type(price))
print("Value:", name, "Type:", type(name))
print("Value:", is_student, "Type:", type(is_student))
```

**Task B2:**

```python
str = "50"

int = int(str)

print("String value:",str, "Type:", type(str))
print("Converted to Integer:", int, "Type:", type(int))
```
